```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import climate as cli
import seaborn as sns

plt.style.use('default')
plt.rcParams.update({'font.size': 16})
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
data = pd.read_csv("DATA111_V0.csv", encoding='latin-1')
data['YEAR'] = data['YEAR'].astype(int)
```

```python
data.head()
```

```python
data['td1'] = cli.get_td(data['N_AIRTMP_MED10'],data['N_AIRMTR_PCT10'])
```

```python
data['h1'] = cli.get_humidex(data['N_AIRTMP_MED10'],data['td1'])
```

```python
data.loc[data['h1'] > 60 , 'h1'] = 50
sns.distplot(data['h1'])
```

```python
# Example 1 dia

year = 2018
day = 133
plt.figure(figsize=(12,5))
for i in data[data.YEAR == year]['DAY365'].unique():
    example = data[(data['YEAR'] == year) & (data['DAY365'] == i)]

    plt.plot(example['MINUTE2400'], example['h1'])

    plt.savefig("fig/HUMIDEX_2018_DIA133")
plt.show()
```
