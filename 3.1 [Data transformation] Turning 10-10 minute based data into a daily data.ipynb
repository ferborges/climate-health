# Creating auxiliar varibles:

- Month
- Week
- Categorical wind direction

## Getting month

```python
def get_month(day):
    if((day>0) & (day <= 31)):
        return 1
    elif((day>31) & (day <= 59)):
        return 2
    elif((day>59) & (day <= 90)):
        return 3
    elif((day>90) & (day <= 120)):
        return 4
    elif((day>120) & (day <= 151)):
        return 5
    elif((day>151) & (day <= 181)):
        return 6
    elif((day>181) & (day <= 212)):
        return 7
    elif((day>212) & (day <= 243)):
        return 8
    elif((day>243) & (day <= 273)):
        return 9
    elif((day>273) & (day <= 304)):
        return 10
    elif((day>304) & (day <= 334)):
        return 11
    else:
        return 12
```

## Getting week

```python
def get_week(days):
    return np.floor(days/7)
```

```python
base['WEEK'] = base['DAY365'].map(get_week)
```

```python
def get_wnddir(x):
    if((x>=348.75) | (x < 11.25)):
        return 'N'
    elif((x>=11.25) & (x < 33.75)):
        return 'NNE'
    elif((x>=33.75) & (x < 56.25)):
        return 'NE'
    elif((x>=56.25) & (x < 78.75)):
        return 'ENE'
    elif((x>=78.75) & (x < 101.25)):
        return 'E'
    elif((x>=101.25) & (x < 123.75)):
        return 'ESE'
    elif((x>=123.75) & (x < 146.25)):
        return 'SE'
    elif((x>=146.25) & (x < 168.75)):
        return 'SSE'
    elif((x>=168.75) & (x < 191.25)):
        return 'S'
    elif((x>=191.25) & (x < 213.75)):
        return 'SSW'
    elif((x>=213.75) & (x < 236.25)):
        return 'SW'
    elif((x>=236.25) & (x < 258.75)):
        return 'WSW'
    elif((x>=258.75) & (x < 281.25)):
        return 'W'
    elif((x>=281.25) & (x < 303.75)):
        return 'WNW'
    elif((x>=303.75) & (x < 326.25)):
        return 'NW'
    elif((x>=326.25) & (x < 348.75)):
        return 'NNW'
```

```python
base['C_WNDDIR_DEGREE'] = base['N_WNDDIR_DEGREE'].map(get_wnddir)
```

# Transforming 10minutos based in a day based data

We are getting max min and respective wind direction for each variable, and the mean of each variable

```python
base['MONTH'] = base['DAY365'].map(get_month)
base['WEEK'] = base['DAY365'].map(get_week)
```

```python
indices = []
# var1 = 'Temperatura do Ar Média (°C)'
# var2 = 'Veloc. Instantânea do Vento a 5 metros (m/s)'
# var3 = 'Veloc. Instantânea do Vento a 2 metros (m/s)'



vardir = 'N_WNDDIR_DEGREE'


# Years to iterate
years = base.YEAR.unique()
years.sort()

# Days to iterate
days = base['DAY365'].unique()
days.sort()

daybase = base

results = []

i = 1

for a in years:
    for d in days:
        # Aux dataset only with information in the respective year and day
        base_aux = base[(base['YEAR'] == a) & (base['DAY365'] == d)]
        
        # Infos is an auxiliar array to save informations to teh day dataset
        infos = []
        infos.append(a)
        infos.append(d)
        
        if(i == 1):
            namevars = []
            namevars.append('YEAR')
            namevars.append('DAY365')
        
        # Do calculations only in years and days that has at least one observation
        if(base_aux.shape[0]>0):    
 
            for v in base.drop(['YEAR','MONTH','DAY365','C_WNDDIR_DEGREE','MINUTE2400','WEEK','COD'], axis = 1).columns.values:
            
                ix_aux = base_aux[base_aux[v] == base_aux[v].max()].index[0]
                infos.append(base_aux.ix[ix_aux][v])
                if(i == 1):
                    str_var = 'MAX_' + v
                    namevars.append(str_var)
                
                infos.append(base_aux.ix[ix_aux][vardir])
                if(i == 1):
                    str_var = 'DIR_MAX_' + v
                    namevars.append(str_var)
                
                ix_aux = base_aux[base_aux[v] == base_aux[v].min()].index[0]
                infos.append(base_aux.ix[ix_aux][v])
                if(i == 1):
                    str_var = 'MIN_' + v
                    namevars.append(str_var)
                
                infos.append(base_aux.ix[ix_aux][vardir])
                if(i == 1):
                    str_var = 'DIR_MIN_' + v
                    namevars.append(str_var)
                
                mean = base_aux[v].mean()
                infos.append(mean)
                if(i == 1):
                    str_var = 'MEAN_' + v
                    namevars.append(str_var)
            
            if(i == 1):
                namevars.append('HUMIDEX')
            
            base_aux['td2'] = cli.get_td(base_aux['N_AIRTMP_MED10'],base_aux['N_AIRMTR_PCT10'])
            base_aux['h2'] = cli.get_humidex(base_aux['N_AIRTMP_MED10'],base_aux['td2'])
            base_aux.loc[base_aux['h2'] > 60 , 'h2'] = 54
            infos.append(base_aux['h2'].max())
            
            results.append(infos)
            i = i + 1

df_result = pd.DataFrame(results, columns=namevars)
```
