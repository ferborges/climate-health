```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from mpl_toolkits import mplot3d

plt.style.use('default')
plt.rcParams.update({'font.size': 16})
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
data = pd.read_csv("DATA111_2DAILY_V1.csv")
data['YEAR'] = data['YEAR'].astype(int)
```

# Analyzing air pression along years

```python
data.columns
```

```python
mean_year = data.groupby('YEAR').agg({'MIN_N_PRESSION_HPA': np.mean, 'MAX_N_PRESSION_HPA': np.mean,'MEAN_N_PRESSION_HPA': np.mean}).reset_index()
```

```python
years = data.YEAR.unique()
years = years.astype(int)

plt.figure(figsize=(12,6))
plt.title('MIN air pression by years')
plt.plot(mean_year['MIN_N_PRESSION_HPA'])
plt.xticks(np.arange(len(years)), years, rotation = 45)
plt.show()

plt.figure(figsize=(12,6))
plt.title('MAX air pression by years')
plt.plot(mean_year['MAX_N_PRESSION_HPA'])
plt.xticks(np.arange(len(years)), years, rotation = 45)
plt.show()

plt.figure(figsize=(12,6))
# plt.title('MEAN air pression by years')
plt.plot(mean_year['MEAN_N_PRESSION_HPA'])
plt.xticks(np.arange(len(years)), years, rotation = 45)
plt.savefig("fig/PRESSAO_ANOS")
plt.show()
```

By the graphs above its clearly that air pression is decreasing along years
